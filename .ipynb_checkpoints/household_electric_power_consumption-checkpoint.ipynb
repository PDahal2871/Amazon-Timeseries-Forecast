```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("household_power_consumption.txt", sep=";")
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print(df.columns.to_list())
```

```python
df[df["Global_active_power"] == '?']  #Checking if ? is present or not
```

```python
#replacing ? with Nan to handle nan values
for col in df.columns.to_list()[2:]:
    df[col] = df[col].replace("?", np.nan)
```

```python
df[df["Global_active_power"] == '?']
```

```python
df.isnull().sum()
```

```python
#forward filling the null values
for col in df.columns.to_list()[2:]:
    df[col] = df[col].fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
for col in df.columns.to_list()[2:]:
    df[col] = df[col].astype('float64')
```

```python
df.info()
```

```python
#Merging both date and time for timestamp
df['Datetime'] = pd.to_datetime(df['Date'] + " "+ df['Time'])
df = df.set_index("Datetime")
```

```python
print(df['Sub_metering_1'].unique())
print()
print(df['Sub_metering_2'].unique())
print()
print(df['Sub_metering_3'].unique())
```

```python
df.head()
```

```python
df.drop(["Date","Time"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Global_active_power'].plot()
```

```python
#Plot for 1 day
plt.figure(figsize=(15,7))
df.loc['2006-12-16 17:24:00': '2006-12-17 17:24:00','Global_active_power'].plot(c="g")
df.loc['2006-12-16 17:24:00': '2006-12-17 17:24:00','Global_reactive_power'].plot(c="r")
plt.legend()
```

```python
#smothing the line by taking average between windows
df_gap = df['Global_active_power'].rolling(60*24).mean()
df_rap = df['Global_reactive_power'].rolling(60*24).mean()
```

```python
plt.figure(figsize=(15,7))
df_gap.plot(c='g')
df_rap.plot(c="r")
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
df.to_csv("household_data.csv", index=False)
```

```python
df1 = mini_df[mini_df['Global_active_power'] <1]
df2 = mini_df[(mini_df['Global_active_power'] <2) & (mini_df['Global_active_power'] >=1)]
df3 = mini_df[(mini_df['Global_active_power'] <3) & (mini_df['Global_active_power'] >=2)]
df4 = mini_df[(mini_df['Global_active_power'] <4) & (mini_df['Global_active_power'] >=3)]
df5 = mini_df[(mini_df['Global_active_power'] <5) & (mini_df['Global_active_power'] >=4)]
df6 = mini_df[(mini_df['Global_active_power'] <6) & (mini_df['Global_active_power'] >=5)]
df7 = mini_df[(mini_df['Global_active_power'] <7) & (mini_df['Global_active_power'] >=6)]
df8 = mini_df[(mini_df['Global_active_power'] <8) & (mini_df['Global_active_power'] >=7)]
df9 = mini_df[(mini_df['Global_active_power'] <9) & (mini_df['Global_active_power'] >=8)]
df10 = mini_df[(mini_df['Global_active_power'] <10) & (mini_df['Global_active_power'] >=9)]
df11 = mini_df[mini_df['Global_active_power'] >=10]
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
print(df5.shape)
print(df6.shape)
print(df7.shape)
print(df8.shape)
print(df9.shape)
print(df10.shape)
print(df11.shape)
```
